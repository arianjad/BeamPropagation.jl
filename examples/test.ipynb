```julia
using Revise, BeamPropagation, Distributions, StatsBase, StaticArrays, PhysicalConstants.CODATA2018, Plots, BenchmarkTools
```

```julia
const h = PlanckConstant.val
const ħ = h / 2π
const λ = @with_unit 626 "nm"
const k = 2π / λ
const m = @with_unit 57 "u"
const Δv = ħ * k / m
;
```

```julia
vz_μ = @with_unit 125 "m/s"
vz_σ = @with_unit 25 "m/s"
vxy_μ = @with_unit 0 "m/s"
vxy_σ = @with_unit 25 "m/s"
exit_radius = @with_unit 4 "mm"
;
```

```julia
const r = (
    Normal(0, exit_radius/2), 
    Normal(0, exit_radius/2), 
    Normal(0, 0)
)
const v = (
    Normal(vxy_μ, vxy_σ), 
    Normal(vxy_μ, vxy_σ), 
    Normal(vz_μ, vz_σ)
)
const a = (
    Normal(0, 0),
    Normal(0, 0), 
    Normal(0, 0)
)
;
```

```julia
const detect_zloc = @with_unit 70 "cm"
const dead_rad = @with_unit 100 "cm"
const dead_len = @with_unit 71 "cm"
;
```

```julia
@inline function simple_prop(r, v)
    dist_detect = detect_zloc - r[3]
    x_final = r[1] + v[1] * dist_detect / v[3]
    y_final = r[2] + v[2] * dist_detect / v[3]
    return sqrt(x_final^2 + y_final^2)
end

@inline discard(r, v) = (simple_prop(r, v) > dead_rad) || (r[3] > dead_len)
;
```

```julia
@inline function update(particles)
    return nothing
end

@inline function f(r, v, p)
    return SVector(1e7, 0.0, 0.0)
end

@inline function save(particles, p, s)
    for i in 1:size(particles, 1)
        idx = particles[i].idx
        push!(s.rs[idx], particles.r[i])
        push!(s.dts[idx], particles.dt[i])
    end
    return nothing
end

# @inline function save(particles, p, s)
#     return nothing
# end
;
```

```julia
n = Int64(1e4)
delete_every = 15
save_every = 10
max_steps = 35000
dt = 1e-5
dt_min = 1e-7
dt_max = 1e-3
use_adaptive = true
abstol = 1e-6
alg = "rkf12"
;
```

```julia
using StructArrays
particles = StructArray{Particle}(undef, n)

rs = [Array{SVector{3, Float64}}(undef, 0) for _ in 1:n]
dts = [Array{Float64}(undef, 0) for _ in 1:n]

p = @params ()
s = @params (rs, dts)
;

@time propagate_particles!(r, v, a, alg, particles, f, save, discard, save_every, delete_every, max_steps, update, p, s, dt, use_adaptive, dt_min, dt_max, abstol)

xs = [r[1] for r in s.rs[1]]
zs = [r[3] for r in s.rs[1]]
;
```
