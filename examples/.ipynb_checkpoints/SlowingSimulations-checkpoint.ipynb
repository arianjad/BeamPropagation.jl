```julia
# using Revise, BeamPropagation, Distributions, StatsBase, StaticArrays, PhysicalConstants.CODATA2018, Plots, BenchmarkTools
using Distributions, PhysicalConstants.CODATA2018, StatsBase, StaticArrays, Plots, BenchmarkTools, DelimitedFiles, Unitful
```

```julia
macro params(fields_tuple)
    fields = fields_tuple.args
    esc(
        quote
            NamedTuple{Tuple($fields)}(($fields_tuple))
        end)
end

macro with_unit(arg1, arg2)
    arg2 = @eval @u_str $arg2
    return convert(Float64, upreferred(eval(arg1) .* arg2).val)
end

function dtstep!(rs, vs, as, dt)
    @inbounds @simd for i in eachindex(rs, vs, as)
        rs[i] += vs[i] * dt + as[i] * (0.5 * dt^2)
        vs[i] += as[i] * dt
    end
    return nothing
end

function update_dead!(rs, vs, dead, is_dead)
    @inbounds @simd for i in eachindex(rs)
        dead[i] = is_dead(rs[i], vs[i])
    end
    return nothing
end

function save!(save, idxs, rs, vs, as, states, s)
    @inbounds for i in 1:size(rs,1)
        save(idxs[i], rs[i], vs[i], as[i], states[i], s)
    end
    return nothing
end

function update!(f, rs, vs, as, states, dt, p, s)
    @inbounds @simd for i in eachindex(rs, vs, as)
        (state′, v′, a) = f(rs[i], vs[i], as[i], states[i], dt, p, s)
        states[i] = state′
        vs[i] = v′
        as[i] = a′
    end
    return nothing
end

function propagate!(n, r, v, a, f, save, discard, delete_every, dt, max_steps, p, s)

    idxs = collect(1:n)
    dead = zeros(Bool, n)
    
    rs = [SVector(0.0, 0.0, 0.0) for _ in 1:n]
    vs = [SVector(0.0, 0.0, 0.0) for _ in 1:n]
    as = [SVector(0.0, 0.0, 0.0) for _ in 1:n]
    states = ones(Int64, n)
    
    @inbounds for i in 1:size(rs, 1)
        rs[i] = SVector.(rand(r[1]), rand(r[2]), rand(r[3]))
        vs[i] = SVector.(rand(v[1]), rand(v[2]), rand(v[3]))
        as[i] = SVector.(0.0, 0.0, 0.0)
    end
        
    step = 0
    save_step = 1
    while (step < max_steps)

        # Save positions and velocities
        save!(save, idxs, rs, vs, as, states, s)

        if step % delete_every == 0
            update_dead!(rs, vs, dead, discard)

            # Positions and velocities for "dead" particles are saved
            dead_rs = view(rs, dead)
            dead_vs = view(vs, dead)
            dead_idxs = idxs[dead]

            deleteat!(rs, dead)
            deleteat!(vs, dead)
            deleteat!(as, dead)
            deleteat!(states, dead)
            deleteat!(idxs, dead)
            deleteat!(dead, dead)
        end
        
        update!(f, rs, vs, as, states, dt, p, s)
        
        dtstep!(rs, vs, as, dt)
        step += 1

    end
    return nothing
end
```

```julia
const h = PlanckConstant.val
const ħ = h / 2π
const λ = @with_unit 626 "nm"
const k = 2π / λ
const m = @with_unit 57 "u"
const Δv = ħ * k / m
;
```

```julia
vz_μ = @with_unit 125 "m/s"
vz_σ = @with_unit 25 "m/s"
vxy_μ = @with_unit 0 "m/s"
vxy_σ = @with_unit 25 "m/s" #approximately 4 K
exit_radius = @with_unit 4 "mm"
;
```

```julia
# Note: are these initial position and velocity distributions okay...I think so since they're Gaussian, but should double check cylindrical symmetry
const r = (
    Normal(0, exit_radius/2), 
    Normal(0, exit_radius/2), 
    Normal(0, 0)
)
const v = (
    Normal(vxy_μ, vxy_σ), 
    Normal(vxy_μ, vxy_σ), 
    Normal(vz_μ, vz_σ)
)
const a = (
    Normal(0, 0), 
    Normal(0, 0), 
    Normal(0, 0)
)
;
```

```julia
VBRs = Weights([
    0.9457,     # to 000
    0.0447,     # 100
    3.9e-3,     # 0200
    2.7e-3,     # 200
    9.9e-4,     # 0220
    11.3e-4,    # 0110
    3.9e-4,     # 1200
    1.5e-4,     # 300
    1.3e-4,     # 1220
    0.7e-4,     # 110 (only N=1, assuming roughly 2/3 to 1/3 rotational branching)
    0.4e-4,     # 110, N=2
    5.7e-5,     # 220
    4.3e-5      # other states
])

addressed = [
    true,   # to 000
    true,   # 100
    true,   # 0200
    true,   # 200
    true,   # 0220
    true,   # 0110
    true,   # 1200
    true,   # 300
    true,   # 1220
    true,   # 110
    false,   # 110, N=2
    false,   # 220
    false   # other states
]

transverse = [
    false,  # to 000
    false,  # 100
    false,  # 0200
    false,  # 200
    false,  # 0220
    false,  # 0110
    false,  # 1200
    false,  # 300
    false,   # 1220
    false,   # 110
    false,   # 110, N=2
    false,   # 220
    false   # other states
]
;
```

```julia
@inline function transverse_on(z)
    if 0.175 < z < 0.20     # 1st transverse region 17.5 - 20 cm after cell
        return true
    elseif 0.45 < z < 0.50  # 2nd transverse region 45 - 50 cm after cell
        return true
    elseif 0.56 < z < 0.61  # 2nd transverse region 56 - 61 cm after cell
        return true
    else
        return false
    end
end

@inline function random_unit3Dvector()
    θ = rand(Uniform(0, 2π))
    z = rand(Uniform(-1, 1))
    return @SVector [sqrt(1-z^2)*cos(θ), sqrt(1-z^2)*sin(θ), z]
end
;
```

```julia
const detect_rad  = @with_unit 0.5 "cm"
const detect_zloc = @with_unit 71 "cm"
const detect_zlen = @with_unit 0.5 "cm"
const dead_rad = @with_unit 1 "cm"
const dead_len = detect_zloc + detect_zlen
;
```

```julia
@inline function simple_prop(r, v)
    dist_detect = detect_zloc - r[3]
    x_final = r[1] + v[1] * dist_detect / v[3]
    y_final = r[2] + v[2] * dist_detect / v[3]
    return sqrt(x_final^2 + y_final^2)
end

@inline discard(r, v) = (simple_prop(r, v) > dead_rad) || (r[3] > dead_len)
@inline is_detectable(r) = sqrt(r[1]^2 + r[2]^2) < detect_rad && (detect_zloc + detect_zlen > r[3] > detect_zloc)
;
```

## Propagating with discrete velocity kicks for each photon (rather than via a mean deceleration)

```julia
@inline function save(i, r, v, a, state, s)
    s.vzs[i] = v[3]
    return nothing
end
;

@inline function f(i, r, v, a, state, dt, p, s)
    if is_detectable(r) && ~s.detectable[i]
        s.detectable[i] = true
    end
    if slowing_on && p.addressed[state] && (~p.transverse[state] | transverse_on(r[3]))
        state′ = sample(1:13, p.VBRs)
        v′ = @SVector [v[1], v[2], v[3] - Δv]
        v′ += Δv .* random_unit3Dvector()
        s.photons[i] += 1
    else
        state′ = state
        v′ = v
    end
    s.states[i] = state′
    return (state′, v′, a)
end
;
```

```julia
n = Int64(5e6)
scattering_rate = @with_unit 1.2 "MHz" #here "MHz" doesn't include a factor of 2pi, so 1.2 MHz is really 1.2e6 /s
delete_every    = 15
dt              = 1 / scattering_rate
max_steps       = Int64(3.5e4)
;
```

```julia
vzs           = zeros(Float64, n)
photons       = zeros(Int64, n)
detectable    = zeros(Bool, n)
states        = ones(Int64, n)
slowing_on = true

p = @params (VBRs, transverse, addressed, slowing_on)
s = @params (vzs, photons, detectable, states)

@time propagate!(n, r, v, a, f, save, discard, delete_every, dt, max_steps, p, s)
;
```

## Propagating with constant deceleration instead

```julia
@inline function save(i, r, v, a, state, s)
    s.vzs[i] = v[3]
    s.detectable[i] = simple_prop(r, v) < detect_rad
    return nothing
end
;

const scattering_rate_const = @with_unit 1.2 "MHz" #here "MHz" doesn't include a factor of 2pi, so 1.2 MHz is really 1.2e6 /s

@inline function f(i, r, v, a, state, dt, p, s)
    if is_detectable(r) && ~s.detectable[i]
        s.detectable[i] = true
    end
    if p.addressed[state] & (~p.transverse[state] | transverse_on(r[3]))
        state′ = sample(1:13, p.VBRs)
        #v′ = @SVector [v[1], v[2], v[3] - Δv]
        v′ = @SVector [v[1], v[2], v[3]]
        v′ += Δv .* random_unit3Dvector()
        a′ = @SVector [0, 0, -ħ * k / m * scattering_rate_const]
        s.photons[i] += 1
    else
        state′ = state
        v′ = v
        a′ = @SVector [0, 0, 0]
    end
    s.vzs[i] = v′[3]
    s.states[i] = state′
    return (state′, v′, a′)
end
;
```

```julia
@inline function transverse_on(z)
    if 0.175 < z < 0.20     # 1st transverse region 17.5 - 20 cm after cell
        return true
    elseif 0.45 < z < 0.50  # 2nd transverse region 45 - 50 cm after cell
        return true
    elseif 0.56 < z < 0.61  # 2nd transverse region 56 - 61 cm after cell
        return true
    else
        return false
    end
end

#is this really uniform? I don't think so...
@inline function random_unit3Dvector()
    θ = rand(Uniform(0, 2π))
    z = rand(Uniform(-1, 1))
    return @SVector [sqrt(1-z^2)*cos(θ), sqrt(1-z^2)*sin(θ), z]
end
;
```

```julia
const detect_rad  = @with_unit 0.5 "cm"
const detect_zloc = @with_unit 71 "cm"
const detect_zlen = @with_unit 0.5 "cm"
const dead_rad = @with_unit 1 "cm"
const dead_len = detect_zloc + detect_zlen
;
```

```julia
@inline function simple_prop(r, v)
    dist_detect = detect_zloc - r[3]
    x_final = r[1] + v[1] * dist_detect / v[3]
    y_final = r[2] + v[2] * dist_detect / v[3]
    return sqrt(x_final^2 + y_final^2)
end

@inline discard(r, v) = (simple_prop(r, v) > dead_rad) || (r[3] > dead_len)
@inline is_detectable(r) = sqrt(r[1]^2 + r[2]^2) < detect_rad && (detect_zloc + detect_zlen > r[3] > detect_zloc)
;
```

```julia
@inline function save(i, r, v, a, state, s)
    s.vzs[i] = v[3]
    s.detectable[i] = simple_prop(r, v) < detect_rad
    return nothing
end
;

const scattering_rate_const = @with_unit 1.2 "MHz" #here "MHz" doesn't include a factor of 2pi, so 1.2 MHz is really 1.2e6 /s

@inline function f(i, r, v, a, state, dt, p, s)
    if is_detectable(r) && ~s.detectable[i]
        s.detectable[i] = true
    end
    if p.addressed[state] & (~p.transverse[state] | transverse_on(r[3]))
        state′ = sample(1:13, p.VBRs)
        #v′ = @SVector [v[1], v[2], v[3] - Δv]
        v′ = @SVector [v[1], v[2], v[3]]
        v′ += Δv .* random_unit3Dvector()
        a′ = @SVector [0, 0, -ħ * k / m * scattering_rate_const]
        s.photons[i] += 1
    else
        state′ = state
        v′ = v
        a′ = @SVector [0, 0, 0]
    end
    s.vzs[i] = v′[3]
    s.states[i] = state′
    return (state′, v′, a′)
end
;
```

```julia
tvec = [0.175, 0.20, 0.45, 0.50, 0.56, 0.61] #start 1, end 1, start 2, end 2, etc.

@inline function transverse_on(z,p)
    if p.tvec[1] < z < p.tvec[2]     # 1st transverse region 17.5 - 20 cm after cell
        return true
    elseif p.tvec[3] < z < p.tvec[4]  # 2nd transverse region 45 - 50 cm after cell
        return true
    elseif p.tvec[5] < z < p.tvec[6]  # 2nd transverse region 56 - 61 cm after cell
        return true
    else
        return false
    end
end

#is this really uniform? I don't think so...
@inline function random_unit3Dvector()
    θ = rand(Uniform(0, 2π))
    z = rand(Uniform(-1, 1))
    return @SVector [sqrt(1-z^2)*cos(θ), sqrt(1-z^2)*sin(θ), z]
end
;
```

```julia
const detect_rad  = @with_unit 0.5 "cm"
const detect_zloc = @with_unit 71 "cm"
const detect_zlen = @with_unit 0.5 "cm"
const dead_rad = @with_unit 1 "cm"
const dead_len = detect_zloc + detect_zlen
;
```

```julia
@inline function simple_prop(r, v)
    dist_detect = detect_zloc - r[3]
    x_final = r[1] + v[1] * dist_detect / v[3]
    y_final = r[2] + v[2] * dist_detect / v[3]
    return sqrt(x_final^2 + y_final^2)
end

@inline discard(r, v) = (simple_prop(r, v) > dead_rad) || (r[3] > dead_len)
@inline is_detectable(r) = sqrt(r[1]^2 + r[2]^2) < detect_rad && (detect_zloc + detect_zlen > r[3] > detect_zloc)
;
```

```julia
@inline function save(i, r, v, a, state, s)
    s.vzs[i] = v[3]
    s.detectable[i] = simple_prop(r, v) < detect_rad
    return nothing
end
;

const scattering_rate_const = @with_unit 1.2 "MHz" #here "MHz" doesn't include a factor of 2pi, so 1.2 MHz is really 1.2e6 /s

@inline function f(i, r, v, a, state, dt, p, s)
    if is_detectable(r) && ~s.detectable[i]
        s.detectable[i] = true
    end
    if p.addressed[state] & (~p.transverse[state] | transverse_on(r[3],p))
        state′ = sample(1:13, p.VBRs)
        #v′ = @SVector [v[1], v[2], v[3] - Δv]
        v′ = @SVector [v[1], v[2], v[3]]
        v′ += Δv .* random_unit3Dvector()
        a′ = @SVector [0, 0, -ħ * k / m * scattering_rate_const]
        s.photons[i] += 1
    else
        state′ = state
        v′ = v
        a′ = @SVector [0, 0, 0]
    end
    s.vzs[i] = v′[3]
    s.states[i] = state′
    return (state′, v′, a′)
end
;
```

```julia
VBRs = Weights([
    0.9457,     # to 000
    0.0447,     # 100
    3.9e-3,     # 0200
    2.7e-3,     # 200
    9.9e-4,     # 0220
    11.3e-4,    # 0110
    3.9e-4,     # 1200
    1.5e-4,     # 300
    1.3e-4,     # 1220
    0.7e-4,     # 110 (only N=1, assuming roughly 2/3 to 1/3 rotational branching)
    0.4e-4,     # 110, N=2
    5.7e-5,     # 220
    4.3e-5      # other states
])

addressed = [
    true,   # to 000
    true,   # 100
    true,   # 0200
    true,   # 200
    true,   # 0220
    true,   # 0110
    true,   # 1200
    true,   # 300
    true,   # 1220
    true,   # 110
    false,   # 110, N=2
    false,   # 220
    false   # other states
]

transverse = [
    false,  # to 000
    false,  # 100
    false,  # 0200
    false,  # 200
    false,  # 0220
    false,  # 0110
    false,  # 1200
    false,  # 300
    true,   # 1220
    true,   # 110
    false,   # 110, N=2
    false,   # 220
    false   # other states
]
;
```

```julia
tvec = [0.175, 0.20, 0.45, 0.50, 0.56, 0.61] #start 1, end 1, start 2, end 2, etc.

@inline function transverse_on(z,p)
    if p.tvec[1] < z < p.tvec[2]     # 1st transverse region 17.5 - 20 cm after cell
        return true
    elseif p.tvec[3] < z < p.tvec[4]  # 2nd transverse region 45 - 50 cm after cell
        return true
    elseif p.tvec[5] < z < p.tvec[6]  # 2nd transverse region 56 - 61 cm after cell
        return true
    else
        return false
    end
end

#is this really uniform? I don't think so...
@inline function random_unit3Dvector()
    θ = rand(Uniform(0, 2π))
    z = rand(Uniform(-1, 1))
    return @SVector [sqrt(1-z^2)*cos(θ), sqrt(1-z^2)*sin(θ), z]
end
;
```

```julia
const detect_rad  = @with_unit 0.5 "cm"
const detect_zloc = @with_unit 60 "cm"
const detect_zlen = @with_unit 0.5 "cm"
const dead_rad = @with_unit 1 "cm"
const dead_len = detect_zloc + detect_zlen
;
```

```julia
@inline function simple_prop(r, v)
    dist_detect = detect_zloc - r[3]
    x_final = r[1] + v[1] * dist_detect / v[3]
    y_final = r[2] + v[2] * dist_detect / v[3]
    return sqrt(x_final^2 + y_final^2)
end

@inline discard(r, v) = (simple_prop(r, v) > dead_rad) || (r[3] > dead_len)
@inline is_detectable(r) = sqrt(r[1]^2 + r[2]^2) < detect_rad && (detect_zloc + detect_zlen > r[3] > detect_zloc)
;
```

```julia
@inline function save(i, r, v, a, state, s)
    s.vzs[i] = v[3]
    s.detectable[i] = simple_prop(r, v) < detect_rad
    return nothing
end
;

const scattering_rate_const = @with_unit 1.2 "MHz" #here "MHz" doesn't include a factor of 2pi, so 1.2 MHz is really 1.2e6 /s

@inline function f(i, r, v, a, state, dt, p, s)
    if is_detectable(r) && ~s.detectable[i]
        s.detectable[i] = true
    end
    if p.addressed[state] & (~p.transverse[state] | transverse_on(r[3],p))
        state′ = sample(1:13, p.VBRs)
        #v′ = @SVector [v[1], v[2], v[3] - Δv]
        v′ = @SVector [v[1], v[2], v[3]]
        v′ += Δv .* random_unit3Dvector()
        a′ = @SVector [0, 0, -ħ * k / m * scattering_rate_const]
        s.photons[i] += 1
    else
        state′ = state
        v′ = v
        a′ = @SVector [0, 0, 0]
    end
    s.vzs[i] = v′[3]
    s.states[i] = state′
    return (state′, v′, a′)
end
;
```

```julia
n = Int64(20e6)
#scattering_rate = @with_unit 1.2 "MHz" #here "MHz" doesn't include a factor of 2pi, so 1.2 MHz is really 1.2e6 /s
delete_every    = 15
dt              = 1 / scattering_rate_const
max_steps       = Int64(3.5e4)
;
```

```julia
vzs           = zeros(Float64, n)
photons       = zeros(Int64, n)
detectable    = zeros(Bool, n)
states        = ones(Int64, n)

p = @params (VBRs, transverse, addressed, tvec)
s = @params (vzs, photons, detectable, states)

@time s0, st = propagate!(n, r, v, a, f, save, discard, delete_every, dt, max_steps, p, s)
;
```

```julia
bright = [addressed[final_state] for final_state in st.states]
;
```

```julia
barhist(s0.vzs[s0.detectable], bins=0:25/4:225, alpha=0.5, xlim=[0,225], label="Unperturbed")
barhist!(st.vzs[st.detectable], bins=0:25/4:225, alpha=0.5, label="Slowed")
barhist!(st.vzs[st.detectable .& bright], bins=0:25/4:225, alpha=0.5, label="Slowed and detected")
xlabel!("Velocity (m/s)")
ylabel!("Molecule count")
```

```julia
v_class = 0 .< st.vzs .< 100
histogram2d(st.vzs[st.detectable .& bright .& v_class], s0.vzs[st.detectable .& bright .& v_class], bins=30)
ylabel!("Initial velocity (m/s)")
xlabel!("Slowed velocity (m/s)")
```

```julia
v_low_cut = 10
v_low = abs.(st.vzs) .< v_low_cut
println("Number detected in unperturbed beam: ", length(s0.vzs[s0.detectable]))
println("Number detectable in slowed beam: ", length(st.vzs[st.detectable]))
println("Number detected (i.e. in bright states) in slowed beam: ", length(st.vzs[st.detectable .& bright]))
println("Number detected below ", v_low_cut, " m/s in slowed beam: ", length(st.vzs[st.detectable .& bright .& v_low]))
slow_frac = length(st.vzs[st.detectable .& bright .& v_low])/length(s0.vzs[s0.detectable])
println("Fraction of unperturbed below ", v_low_cut, " m/s: ", slow_frac)
println("Number slowed from 1e7 initial: ", slow_frac*1e7)
```
